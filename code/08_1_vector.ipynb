```python
import numpy as np
import utils.dataset as dataset
import utils.plot_utils as plot_utils
# %matplotlib
%matplotlib ipympl
```

```python
## Create a dataset
n = 100
X, Y = dataset.get_beans5(n)

# 输入xs是二维向量
# print(X) 
# print(Y)
```

```python
plot_utils.show_scatter(X,Y)
```

```python


W = np.array([0.1,0.1])
B = np.array([0.1])


def forward_propgation(X):
    Z= X.dot(W.T)+B
    A = 1/(1+np.exp(-Z))
    return A


plot_utils.show_scatter_surface(X,Y, forward_propgation)
```

```python
for _ in range(500):
    for i in range(n):
        Xi = X[i]
        Yi = Y[i]

        # 前向传播
        A = forward_propgation(Xi)
        E = (Yi-A)**2
        # 反向传播
        dEdA = -2*(Yi-A)
        dAdZ = A*(1-A)
        dZdW = Xi
        dZdB = 1

        dEdW = dEdA*dAdZ*dZdW
        dEdB = dEdA*dAdZ*dZdB

        # 梯度下降
        alpha = 0.01
        W = W - alpha*dEdW
        B = B - alpha*dEdB
plot_utils.show_scatter_surface(X, Y, forward_propgation)
```
